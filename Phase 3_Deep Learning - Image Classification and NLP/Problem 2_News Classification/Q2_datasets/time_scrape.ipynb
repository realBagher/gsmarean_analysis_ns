```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import json
import re
import time

headers={'authority': 'www.google.com',
    'accept-language': 'en-US,en;q=0.9',
    'cache-control': 'max-age=0',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
```

```python
# Sport
headlines_s=[]
descriptions_s=[]
labels_s=[]
for i in range(1,233):

  print("----------page "+str(i)+"-------------")
  page = requests.get('https://time.com/section/sports/'+'?page='+str(i),headers=headers)
  soup = BeautifulSoup(page.content, 'html.parser')
  cards=soup.find('div',class_='component taxonomy-related-touts section-related__touts')
  head=cards.find_all('h2',class_="headline")
  for a in head:
    headlines_s.append(a.text)

  for j in cards.find_all('a', href=True):

    new_ref=j["href"]
    print(new_ref)

    p = requests.get("https://www.time.com"+new_ref,headers=headers)
    s = BeautifulSoup(p.content, 'html.parser')


    try:
      descriptions_s.append(soup.find('div',id="article-body-main").text)
    except:
      try:
        par=""
        body=s.find('div',id="article-body-main")
        for x in body.find_all("p",class_="self-baseline px-0 font-pt-serif text-17px leading-7 tracking-0.5px"):
          par+=x.text+" "
        descriptions_s.append(par)
      except:
        descriptions_s.append("None")
    labels_s.append("Sport")

  if i%10==0:
    time.sleep(20)
```

```python
# sports dataframe
import pandas as pd
sport={"headline":headlines_s,"description":descriptions_s,"label":labels_s}
sports=pd.DataFrame(sport)
```

```python
sports.to_excel("sports.xlsx")
```
